```python
import numpy as np 
import random

train_data = np.array([[1.0,1.0]])
train_targets = np.array([1.0])

test_data = np.array([[2.0,2.0]])
test_targets = np.array([4.0])

for i in range(1,10000):
    n1 = random.randint(1, 100)
    n2 = random.randint(1, 100)
    train_data= np.append(train_data,[[n1,n2]],axis=0)
    train_targets= np.append(train_targets,[n1*n2])
    
    test_data = np.append(test_data,[[n1,n2]],axis=0)
    test_targets = np.append(test_targets,[n1*n2])

for i in range(1,2500):
    n1 = random.randint(1, 50)
    n2 = random.randint(1, 50)
    train_data= np.append(train_data,[[n1,n2]],axis=0)
    train_targets= np.append(train_targets,[n1*n2])
    
    test_data = np.append(test_data,[[n1,n2]],axis=0)
    test_targets = np.append(test_targets,[n1*n2])

for i in range(1,2500):
    n1 = random.randint(50, 100)
    n2 = random.randint(50, 100)
    train_data= np.append(train_data,[[n1,n2]],axis=0)
    train_targets= np.append(train_targets,[n1*n2])
    
    test_data = np.append(test_data,[[n1,n2]],axis=0)
    test_targets = np.append(test_targets,[n1*n2])
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(2,)),
    keras.layers.Dense(20, activation=tf.nn.relu),
	keras.layers.Dense(20, activation=tf.nn.relu),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse',metrics=['mae'])
model.fit(train_data, train_targets, epochs=15, batch_size=1)
test_loss, test_accuracy = model.evaluate(test_data, test_targets)

print('Test accuracy:', test_accuracy)
print('Test results:')
a= np.array([[30,20],[50,20],[10,10], [7,8],[7,20],[30,70]])
print(model.predict(a))
```
